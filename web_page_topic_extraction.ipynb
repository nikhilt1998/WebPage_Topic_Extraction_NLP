Natural Language Processing using Python!!!

Extract topic of websites such as Wikipedia!!!

Importing NLTK 

```python
import nltk
nltk.download()
```

Importing urlib to crawl the webpage

```python
import urllib.request
response =  urllib.request.urlopen('https://en.wikipedia.org/wiki/Python_(programming_language)')
html = response.read()
print(html)
```

Import BeautifulSoup for pulling data out of HTML and XML files

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html,'html5lib')
text = soup.get_text(strip = True)
print(text)
```

Convert the text into tokens.

```python
tokens = [t for t in text.split()]
print(tokens)
```

Count word Frequency

```python
from nltk.corpus import stopwords
sr= stopwords.words('english')
clean_tokens = tokens[:]
for token in tokens:
    if token in stopwords.words('english'):
        
        clean_tokens.remove(token)
freq = nltk.FreqDist(clean_tokens)
for key,val in freq.items():
    print(str(key) + ':' + str(val))
```

Frequency plot to check maximum occurrences of a word i.e. Topic

```python
freq.plot(20, cumulative=False)
```

The Topic is : Python!!!
